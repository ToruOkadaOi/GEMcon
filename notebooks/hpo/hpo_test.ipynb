```python

# 2. Get a specific term by ID
print("\n2. Getting details for HP:0001945...")
term = details.sync(client=client, id="HP:0001945")
print(f"   Term: {term}")
```

```python

# 3. Get parents
print("\n3. Getting parents of HP:0001250...")
parent_list = parents.sync(client=client, id="HP:0001250")
print(f"   Found {len(parent_list)} parent(s)")
print(parent_list)
```

```python

# 4. Get children
print("\n4. Getting children of HP:0001250...")
child_list = children.sync(client=client, id="HP:0001250")
print(f"   Found {len(child_list)} children")
print(child_list)
```

```python
# 5. Get ancestors
print("\n5. Getting ancestors of HP:0001250...")
ancestor_list = ancestors.sync(client=client, id="HP:0001250")
print(f"   Found {len(ancestor_list)} ancestors")
print(ancestor_list)
```

```python

# 6. Get descendants
print("\n6. Getting descendants of HP:0001250...")
descendant_list = descendants.sync(client=client, id="HP:0001250")
print(f"   Found {len(descendant_list)} descendants")
print(descendant_list)
```
