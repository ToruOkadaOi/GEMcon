```python
import pandas as pd

# Load your data
expr = pd.read_csv("expression_data.csv", index_col=0)
ref = pd.read_csv("gtf_reference_genesymbols_ensemblid.tsv", sep="\t")

# Make a mapping dictionary from gene_name → gene_id
mapping = dict(zip(ref["gene_name"], ref["gene_id"]))

# Map symbols to Ensembl IDs
expr.index = expr.index.map(mapping)

# Drop any genes not found in mapping
expr = expr.dropna().copy()

# Remove version numbers (e.g. ENSG00000147576.12 → ENSG00000147576)
expr.index = expr.index.str.replace(r"\.\d+$", "", regex=True)

# Drop duplicates if any
expr = expr[~expr.index.duplicated(keep="first")]
```
